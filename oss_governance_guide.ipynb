# 오픈소스 기여 전략 수립을 위한 고려사항
	* 오픈소스 개발 기여가 중요한 이유
		* 경영진을 설득하고 적극적인 지원을 받기 위해 필요
	* 현재 기업이 전략적으로 사용하는 오픈소스
	* 기업의 비즈니스와 제품에 큰 영향을 미치는 중요한 프로젝트
	* 전문가 확보 여부
		* 필요하다면 자원과 예산 확보를 해야 함
	* 지원 자금 필요여부
		* 프로젝트 지원을 위한 기업 멤버쉽 및 후원
		* 콘퍼런스 및 행사 후원
	* 홍보 방법
		* 컨퍼런스 및 이벤트 후원, 세션발표
		* 기업내 구성원의 커뮤니티 활동 참여를 허용하고 적극 권장
	* 기여 가이드와 프로세스 준비
		* 컴플라이언스 및 보안 등 리스크를 최소화할 수 있는 기여 정책과 가이드를 제공
		* 신규 기여자가 법적, 기술적 피드백을 받을 수 있는 내부 검토 프로세스 제공
	* 충분한 교육
		* 라이선스, 거버넌스, 커뮤니티 참여, 기여 모범사례 등 교육을 제공
		* 숙련된 기여자를 멘토로 지원하는 멘토링 프로그램을 통해 신규 기여자 양성

# 오픈소스 기여 정책의 방향을 설정
	* 개방 또는 보수의 방향설정이 필요
	* 개방적 기여정책 예시
		* 우리는 [기업] 외부의 오픈소스SW (및 유사) 프로젝트에 참여할 것을 권장한다. 프로젝트에 참여하고 싶은 경우 자세한 내용은 오픈소스SW 컴플라이언스 책임자에게 문의할 수 있다. [기업]이 관여하는 프로젝트 외에도 자신의 시간에 다른 프로젝트에 자유롭게 기여할 수 있다. 그러나 [기업]의 영업 비밀 및 지적 재산 보호 의무가 있음에 유의해야 한다. 따라서 [기업]에서의 업무와 관련이 있을 수 있는 오픈소스SW 프로젝트에 기여하기 전에 오픈소스SW 컴플라이언스 책임자와 문제 될 만한 여지를 모두 제거 한다.
	* 보수적 기여정책 예시
		* 일반적으로 [기업] 외부의 오픈소스SW (및 유사) 프로젝트에 참여하는 것을 권장하지 않는다. [기업] 외부 프로젝트에 참여하고 싶다면 오픈소스SW 연락 담당자에게 명시적인 허가를 요청해야 한다. 이것은 [기업]의 영업 비밀 및 지적 재산 보호 의무를 고려할 때 필요하다.

# 오픈소스 기여정책 포함 사항
	* 기여를 허용하는 기준 수립
		* 기업의 오픈소스 책임자가 내부 기여자들의 외부 오픈소스 프로젝트 기여에 대한 점검 또는 구성원들의 판단하에 자율적인 기여인지
		* 자율적인 기여 상황 예시
			* 100 라인 이하의 작은 코드 조각
			* Stack Overflow 문의/ 답변
			* Bug Report
			* GitHub 활동 : Pull Request 수락, Comment 대응, 버전 릴리즈 등
		* 기업의 리뷰와 승인을 거치는 상황 예시
			* 상당한 양의 기여
			* 라이선스 정책으로 인해 기업이 사용할 수 없는 프로젝트로의 기여
			(예: AGPL, CC BY-NC*, 상업적 사용을 제한하는 프로젝트 등)
			* Private Repository (단, 기업과 별도의 서면 약정이 체결된 경우는 예외)
		* 기여정책 예시
			* 기여할 권리가 있는 코드만 기여하라. 즉, 직접 작성한 코드를 기여한다.
			* 민감한 정보, 특허 등 기업의 지식재산 노출이 우려되는 코드, 문서 기여하지 마라.
			* 가능한 프로젝트의 기여 프로세스 (CONTRIBUTING.md) 에서 요구하는 바를 준수한다.
			* 수준 이하의 코드는 기여하지 마라. 이는 기업의 평판에도 영향을 미칠 수 있다.
			* 이외 어느 것이라도 무언가 바람직하지 않아 보이는 상황이 있다면 주저하지 말고 OSPO에 문의한다.
	* 업무시간 외 개발한 기여코드에 대한 정책
		* 근로시간 외라도 업무와 관련있는 코드라면 기업이 저작권을 소유하게 될 수 도 있다
		* 지식재산 유출 사고 방지를 위해 정책적으로 구성원에게 오픈소스 프로젝트에 최초 기여시 OSPO의 리뷰와 승인을 한차례 받도록 한다
	* 기업계정의 이메일 사용
		* 기업대표하여 커뮤니티와 커뮤니케이션에 대한 책임감을 부여
		* 기업에 대한 인지도 개선
	* CLA/DCO 주의
		* 어떤 오픈소스SW 프로젝트는 모든 기여자에게 CLA (Contributor License Agreement)에 서명할 것을 요구
		* 오픈소스SW 프로젝트에서 기여자에게 서명을 요구하는 CLA를 검토하여 기업의 지식재산이 예기치 않게 노출되는 상황이 발생하지 않도록 주의
		* 서명전에 오픈소스 책임자를 통한 리뷰과정 의무화가 필요
		* DCO : 번거로운 CLA를 회피하기 위해 별도 서명 요구 없이 Git환경에서 --signoff 플래그를 사용하여 자신의 기여에 서명하는 단순방식
			* 깃에서만 사용 가능
	* 기업저작권 표시
		* 구성원이 재직기간에 생성한 저작물의 지식재산은 기업이 소유하기에 프로젝트 코드를 기여할 때 기업의 저작권을 표기
		* 단순 버그 수정등의 목적은 저작권 주장이 쉽지 않기에 저작권 표기도 강제화 하지 않음
	* 행동강령 준수
		* 해당 오픈소스 프로젝트의 행동강령을 잘 숙지하고 준수하도록 하고 문제가 발생하면 즉시 OSP에 알리도록 해야 함
			* 기업의 평판을 떨어뜨리는 상황을 미연에 방지
	* 기여자 목록 확보 및 활용
		* 기여자 목록은 귀중한 자산
			* 동일한 오픈소스 프로젝트 기여자들이 기업내에서 기술공유 및 상호발전을 도모
			* 특정 오픈소스 프로젝트에서 이슈가 발생하면 사내 기여자들을 통해 이슈해결을 위한 컨설팅 요청

# 오픈소스 기여전 주의사항
	* 프로젝트 문서 숙지
		* 코딩 스타일, 운영 등 다양한 가이드라인(Contributing, Readme 등)을 숙지하여 해당프로젝트에 적합한 구성원이 되어야 한다
	* CLA/DCO 서명은 주의
		* 복잡한 저작권 침해 이슈에 휘말리는 것을 미연에 방지하기 위해 내부 OSPO 혹은 법무조직에 문의와 검토가 필요
		* CLA Bot은 간편하기에 주의해야 한다
	* 기여코드를 제출하기 전에 점검
		* 기여할 권리가 있는 코드인가? 만약 기여하려는 코드에 직접 작성하지 않은 코드가 포함되었다면 여러분에게 기여할 권리가 있는지 확인해야 한다.
		* 기업의 민감하거나 독점적인 정보가 노출될 수 있는 코드는 제외하라.
		* 기여하려는 코드에 특허가 포함되어 있는지 확인하라. 오픈소스SW 라이선스는 특허 라이선스까지 부여할 것을 요구하기도 한다. 이것이 기업의 정책과 충돌되지 않는지 사전에 확인해야 한다.
		* 스스로 확인하기 어렵거나 모호한 부분이 있다면 기업의 OSPO에 문의한다. OSPO가 없다면 법무 조직이나 상사에게 문의한다.
	* 기업 내부 승인 절차를 진행
		* 저작권적인 측면에서 확인절차가 필요

# 오픈소스 기여 방법
	* 제출하려는 기여코드가 이전에 다뤄진 이력이 있는지 확인
		* README, 이슈, 메일링 리스트 확인
		* 키워드로 검색
	* 이슈  또는 Pull Request를 통해 커뮤니케이션 진행하기에 앞서 관련 가이드 문서 확인
		* 특정 템플릿을 따르거나 사전테스트 요구가 있을 수도 있다
		* 이슈를 사전에 오픈해서 커뮤니티 구성원에게 작업 진행방향을 알리는 것도 좋은 방법
			* 불필요한 작업진행을 안해도 된다
	* 이슈 생성
	* 커뮤니케이션 시작
		* 정확한 의사전달
		* 무조건 물어보기 보다 먼저 스스로 해볼 수 있는 것을 해야 한다
		* 간결한 대화
		* 공개 커뮤니케이션
			* 다른 사람들도 배울 수 있다
		* 문의하고 기다리기
		* 커뮤니티의 결정 존중
		* 품위 유지
			* 글로서 의사소통을 보통 진행하므로 의사소통에 따른 오해가 발생하지 않도록 품위를 유지
	* Pull Request 제출
		* 기여코드 준비되어 있다면 Pull Request로 제출
		* 가능하면 작업이 완료하기 전에 미리 생성하여 피드백을 받는 것이 좋음
			* 제목에 WIP라고 표시
	* 피드백 받기
		* 응답이 없으면 한 일주일 정도 기다려보고 동일한 스레드에 누군가에게 검토 요청
			* @을 통해 맨션 기능 사용
			* 개인적인 연락은 금지
		* 수정요청
			* 빠른 응답을 권장
			* 문제를 더 이상 해결할 여건이 아니라면 메인테이너에게 알려줘야 다른사람이 인수하여 추가 진행이 가능
		* 거절됐다면 그들의 결정을 존중하고 논쟁하거나 적대적인 행동을 하지마라
